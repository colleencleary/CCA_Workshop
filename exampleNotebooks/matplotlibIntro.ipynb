# Plotting data (i.e. matplotlib)

Python doesn't have it's own plotting software, so we're going to introduce the most commonly imported one, Matplotlib.

It's a huge library, very poweful, if a bit of a mess. We're going to introduce a couple of the most useful ones for this kind of data.

...but first we have to import it. This takes a couple more lines than numpy, because it has a few particularly useful sub-libraries, and we need to make it display in the notebook

```python
import matplotlib # the whole package
import matplotlib.pyplot as plt # a useful shortcut to the most common library
from matplotlib import gridspec # a particular tool we'll use later
# and tells the jupyter notebook to show all the plots in the browser
%matplotlib inline

# also we'll use numpy too
import numpy as np
```

```python
# let's make some data 
x = np.linspace(0,10,100)
y = x**2 + 5
```

### The plot command
The simplest matplotlib command, takes x and y coordinates and makes a line plot.

```python
linePlot=plt.gca()
linePlot.plot(x,y) # the simplest plot, in blue
linePlot.plot(x,2*y,c='orange') # in orange
linePlot.plot(x,0.5*y,c='green',lw=3,ls='dotted') # thick, dotted and green
plt.show(linePlot) # displays the plot
```

### Other line plots

There's a whole bunch of variations on simple plots. For example log plots, which are really useful when the range of data is large. We'll only show one example here (semilogy), but there's the equivalent command for a logarithmic x-axis (semilogx), or with two log axes (loglog)

Also, here we show some of the ways you can neaten up your plots (like labelling the axes and setting their ranges)

```python
logPlot=plt.gca()
logPlot.semilogy(x,y, lw =3) # logged y axis
logPlot.set_xlabel('x') # x and y axis labels
logPlot.set_ylabel('y')
logPlot.set_xlim(0,10) # the range of x and y values
logPlot.set_ylim(1,200)
plt.show(logPlot)
```

# Showing 2D data

There's a bunch of ways to show 2D data but we're just going to introduce 1 here today.

IMSHOW shows the values of elements in an array, for example:

```python
data=1-2*np.random.random((10,10)) # a grid of random numbers between -1 and 1
image=plt.gca()
image.imshow(data)
plt.show(image)
```

# Making multiple plots

Often we want more than one plot together, for this we're going to use the tool GRIDSPEC, literally a grid of plots

(We'll also use this as an excuse to show the dangers of not defining the limits of your color bars sensibly!)

```python
grid=gridspec.GridSpec(2,2) # a two by two grid for plots

#top left plot, positive values
posPlot= plt.subplot(grid[0,0])
posPlot.imshow(data,vmin=0,vmax=1) #coloring only differentiates between 0 and 1

#top right plot, negative values
negPlot= plt.subplot(grid[0,1])
negPlot.imshow(data,vmin=-1,vmax=0) #coloring only differentiates between -1 and 0

#bottom left plot, values over a narrow range
narrowPlot= plt.subplot(grid[1,0])
narrowPlot.imshow(data,vmin=-0.5,vmax=0.5) #coloring only differentiates between -0.5 and 0.5

#bottom right plot, values over a wide range
widePlot= plt.subplot(grid[1,1])
widePlot.imshow(data,vmin=-2,vmax=2) #coloring only differentiates between -2 and 2

plt.show(grid)
```
